Ahora bien, para realizar un entrenamiento de datos es necesario quitar los datos faltantes, de la BD que estamos entrenando.
Por ello vamos aponer 0 en donde hay datos nulos.

```python
dataKansas = dataKansas.fillna(0)
print("Tabla con datos eliminados")
dataKansas
```

```python
#Analisis estadistico descriptivo de los datos de nuestra BD
dataKansas.describe()
```

### *Entrenar el modelo*

```python
#importar liberias 
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import Ridge
from sklearn import metrics
```

```python
#Se va escoger o bien establecer las viariables de X , Y
x = dataKansas[['AWND', 'PRCP', 'SNWD']]
y = dataKansas['TMAX']
```

```python
#Ahora se divide los datos para entrenamiento y para pruebas 
#test_size = 0.2 esto es para usarlos como datos de prueba
# el 0.80 restantes es para el entrenamiento
#random_state  este parametro es para controlar la dicivion de los datos
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2,
                                                    random_state = 1234)
```

*Implemertar nuestro modelo lineal* 
# Regresion lineal 

```python
#ajustamos nuestro modelo son regrecion multiple
#ya que utilizamos varias variables para el entrenamiento
mlr = LinearRegression()
mlr.fit(x_train, y_train)
```

```python
#Ahora aplicamos la ecuacion del modelo 
# el cual es la interseccion y el coeficiente
print("Interseccion: ", mlr.intercept_)
print("Coeficiente:")
list(zip(x, mlr.coef_))
```

**Analisis**


```python
#formula para la predicion de nuestraspruebas

mlr_prediccion = mlr.predict(x_test)

#imprimir los datos de la prediicion 

print("Datos predichos: {}".format(mlr_prediccion))
```

```python
#ahora vamos acomparar los valores reales con los predichos
mlr_diff = pd.DataFrame({'Real': y_test,
                         'Predicho': mlr_prediccion})
mlr_diff.head()
```

```python
#ahora proseguimos a evaluar el modelo
from sklearn import metrics

meanAbErr = metrics.mean_absolute_error(y_test, mlr_prediccion)
meanSqErr = metrics.mean_squared_error(y_test, mlr_prediccion)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, mlr_prediccion))

print('R cuadrado: {:.2f}'.format(mlr.score(x,y)*100))
print('Error absoluto medio', meanAbErr)
print('Error cuadrado medio:', meanSqErr)
print('Error cuadrático medio:', rootMeanSqErr)
```

```python
#ahora es importante conover el error general de nuestro
#entrenamiento de datos, de una manera mas precisa
from sklearn.metrics import mean_squared_error

rmse = mean_squared_error(
        y_true = y_test,
        y_pred = mlr_prediccion,
        squared = False
       )
rmse
```
